```python
probability = [0.01, 0.02,0.05,0.1,0.2,0.5]
def addNoise(img: Image.Image, probability: list)->Image.Image:
    """"""
    img = np.array(img)
    for prob in probability:
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                if prob < np.random.random(size=1):
                    img[i,j] = np.random.randint(0, 255)
    return Image.fromarray(img)
```
